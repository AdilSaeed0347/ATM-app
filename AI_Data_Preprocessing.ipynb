No we look data distribution for Each text column

```python
pd.concat([df['customer_id'].value_counts()/df.shape[0] *100,
df3_drop_rows['customer_id'].value_counts()/df3_drop_rows.shape[0] *100], axis=1,
          keys=['customer_original', 'customer_cleaned'])
```

```python
def Cat_var_dist(var):
  return pd.concat([df[var].value_counts()/df.shape[0] *100,
         df3_drop_rows[var].value_counts()/df3_drop_rows.shape[0] *100], axis=1,
         keys=[var+'_original', var+'_cleaned'])

```

```python
Cat_var_dist('customer_id')
```

```python
Cat_var_dist('transaction_date')
```

```python
# Strip leading/trailing spaces from column names
df.columns = df.columns.str.strip()


```

```python
for i in df.columns:
    print(f"Distribution for {i}:")
    try:
        print(Cat_var_dist(i))
    except KeyError as e:
        print(f"Error processing column '{i}': {e}")
```
